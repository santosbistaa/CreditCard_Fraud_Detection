```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data=pd.read_csv('creditcard.csv')
```

```python
#Since,the data is highly unbalanced.So,distribution of legit and fraudulent transactions
credit_card_data['Class'].value_counts()
```

```python
#This dataset is highly unbalanced 0-->Legit and 1-->Fraud 
#Separating the data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comparing the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
#UNDERSAMPLING OR RANDOM SAMPLING:
#Building a similar dataset containing similar distribution of normal transactions and fraudulent transaction
#Since the number of fraudulent transaction is 492
legit_sample=legit.sample(n=492)
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2,stratify=y)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy score on trainig data:",training_data_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy score on test data:",test_data_accuracy)
```
